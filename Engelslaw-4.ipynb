```python
import pandas as pd
import numpy as np
```

```python
expdata1 = pd.read_stata("sec_i.dta")
expdata2 = pd.read_stata("sec_j1.dta")
expdata3 = pd.read_stata("sec_l.dta")
```

```python
expdata1 = expdata1[["case_id","dist","ea","i0a","i05"]]
expdata2 = expdata2[["case_id","j02a","j03a"]]
otherexp = expdata2.loc[lambda expdata2:expdata2["case_id"] == "10101002025","j03a"].sum()
foodexp = expdata1.loc[lambda expdata1:expdata1["case_id"] == "10101002025","i05"].sum()
totexp = foodexp + otherexp
engelcoefficient = foodexp/totexp
engelcoefficient
```

```python
expdata1 = expdata1[["case_id","i05"]]
newfoodexp = expdata1.groupby(["case_id"]).sum()
newnonfoodexp = expdata2.groupby(["case_id"]).sum()
data1 = pd.merge(newfoodexp, newnonfoodexp, on='case_id')
data1["engelcodata"] = data1["i05"]/(data1["i05"]+data1["j03a"])
data1
```

```python
expyear = expdata3[["case_id","l0a","l03"]]
newfoodexp = expdata1.groupby(["case_id"]).sum()
newnonfoodexp1 = expyear.groupby(["case_id"]).sum()
#We multiply i05 by 52.143 because there are 52.143 weeks in a year.
newfoodexp["foodexpyear"] = newfoodexp["i05"].apply(lambda x: x*52.143)
newnonfoodexp["nonfoodexpyear"] = newnonfoodexp["j03a"].apply(lambda x: x*52.143)
totexp1 = pd.merge(newnonfoodexp1, newfoodexp, on='case_id')
totexp1 = pd.merge(totexp1, newnonfoodexp, on='case_id')
totexp1["engelcodata"] = totexp1["foodexpyear"]/(totexp1["foodexpyear"]+totexp1["l03"]+totexp1["nonfoodexpyear"])
totexp1 = totexp1.drop(["l03","i05","j03a","nonfoodexpyear"], axis=1, inplace=True)
totexp1
```
